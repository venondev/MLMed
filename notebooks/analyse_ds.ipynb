```python
import os
import h5py
import numpy as np
import pandas as pd
from scipy import ndimage as ndi

datapath = "./data/train_half"
files = os.listdir(datapath)

ret = []
for idx, file in enumerate(files):
    print(f"{idx}/{len(files)}: {file}")
    with h5py.File(os.path.join(datapath, file), 'r') as f:
        raw = f['raw'][:]
        label = f['label'][:]

        label_labels, num_labels = ndi.label(label)

        unique_aneus = np.unique(label_labels)
        print("unique_aneus in this set", unique_aneus)

        label_volume = np.zeros(num_labels)
        for i in range(1, num_labels + 1):
            m = label_labels[label_labels == i]

            volume = m.sum()

            t = np.zeros(label.shape)
            t[label_labels == i] = 1

            label_volume[i - 1] = volume

            # Get Aneu Size

            x_s = t.sum(axis=(1, 2))
            y_s = t.sum(axis=(0, 2))
            z_s = t.sum(axis=(0, 1))

            x = np.where(x_s)[0][[0, -1]]
            y = np.where(y_s)[0][[0, -1]]
            z = np.where(z_s)[0][[0, -1]]

            ret.append((file, i, volume, x[0], 128 - x[1], y[0], 128 - y[1], z[0], 128 - z[1]))
```

```python
df = pd.DataFrame(ret, columns=["file", "label", "volume", "x_s", "x_e", "y_s", "y_e", "z_s", "z_e"])
df.describe()
```
